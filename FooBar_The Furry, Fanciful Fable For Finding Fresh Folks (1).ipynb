###Problem 3: 

This problem was a recursive arithmetic path problem, wherein a given integer value (as type:str()) has to be reduced to 1 via addition by 1, subtraction by 1, or division by 2 (if even). You have to return a value equal to the number of steps that moves the value to 1. There was some interesting math in the background, which I worked out on paper, but the algorithm itself was fairly simple.

```python
def solution(n):
    
    N = int(n)
    
    if N == 1:
        return 0
    
    Nprime = N
    Steps = 0
    while Nprime != 1:
        if Nprime == 3:
            Nprime += -1
            Steps += 1
        elif Nprime % 2 == 0:
            Nprime = Nprime/2
            Steps +=1
        elif (Nprime+1) % 4 != 0 and (Nprime+1) % 8 !=0:
            Nprime += -1
            Steps+=1
        else:
            Nprime += 1
            Steps+=1
                
    return Steps
```

With some assistance: 

```python
def solution(n):
    
    N = int(n)
    indexAid = N + 1
    
    nullList= []
    for i in range(indexAid):
        sublist = []
        for j in range(indexAid):
            sublist.append(0)
        nullList.append(sublist)
    
    nullList[0][0] = 1
    
    for lst in range(1, len(nullList)):
        for element in range(0, len(nullList)):
            nullList[lst][element] = nullList[lst-1][element]
            if element >= lst:
                nullList[lst][element] += nullList[lst - 1][element-lst]
            
    return (nullList[N][N] - 1)

solution(200)
```
